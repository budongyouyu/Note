# 感知机类

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义感知机模型类
class myPerceptorn:
    
    def __init__(self):
        self.count = 1
        self.w = None
        self.b = 0
        self.l_late = 1
    
    # 算法主函数
    def fit(self,x_trian,y_trian):
        cnt=0
        self.w=np.zeros(x_trian.shape[1])#shape 方法通常用于获取数组或矩阵的形状（维度），主要用于 NumPy 库中的数组对象。
        i = 0

        while i < x_trian.shape[0]:
            x = x_trian[i]
            y = y_trian[i]
            # 判断其所有点是否都没有误分类，如有更新w,b,重新判断
            if y*(np.dot(self.w,x)+self.b)<=0:
                self.w = self.w + self.l_late*np.dot(x,y)
                self.b = self.b+self.l_late*y
                i = 0
            else:
                i += 1
                
            cnt=cnt+1
    
        print("trainning finish ! trainning cost times : {}".format(cnt))
            
```

```python
# 训练集
x_trian = np.array([[3,3],[3,5],[4,7],[4,0],[2,1],[6,2]])
y_trian = np.array([1,1,1,-1,-1,-1])
print(x_trian.size,y_trian.size)
```

```python
# 调用
perceptorn = myPerceptorn()
perceptorn.fit(x_trian,y_trian)

#输出结果
print(perceptorn.w,perceptorn.b,perceptorn.count)
```

# Logic training

```python
# 定义输入数据
x = np.array([[0,0],[0,1],[1,0],[1,1]])

#定义标签值
y_and = np.array([-1,-1,-1,1])
y_or  = np.array([-1,1,1,1])
```

```python
# 例化感知机
logic_and = myPerceptorn()
```

```python
logic_and.fit(x,y_and)
```

```python
# 显示输入输出的二维分布
plt.scatter(x[0:3,0],x[0:3,1],marker='+', color='orange',label='postive')
plt.scatter(x[3:4,0],x[3:4,1],marker='+', color='green',label='postive')

# 模型方程
x_plot = np.linspace(0, 1.5 ,10)
y_plot = -( logic_and.w[0] * x_plot + logic_and.b )/logic_and.w[1]

# 图示
plt.plot(x_plot,y_plot)

#显示
plt.show()
```

```python
logic_or = myPerceptorn()
logic_or.fit(x,y_or)

# 显示输入输出的二维分布
plt.scatter(x[0:1,0],x[0:1,1],marker='*', color='orange',label='postive')
plt.scatter(x[1:4,0],x[1:4,1],marker='*', color='green',label='postive')

# 模型方程
x_plot = np.linspace(-0.5, 1 ,10)
y_plot = -( logic_or.w[0] * x_plot + logic_or.b )/logic_or.w[1]

# 图示
plt.plot(x_plot,y_plot)

#显示
plt.show()
```
